# 02 Exploratory Data Analysis
Professional EDA notebook for Smart Retail Analytics.

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

customers=pd.read_csv('../Dataset/customers.csv')
orders=pd.read_csv('../Dataset/orders.csv')
products=pd.read_csv('../Dataset/products.csv')
inventory=pd.read_csv('../Dataset/inventory.csv')
targets=pd.read_csv('../Dataset/sales_targets.csv')

products.columns=products.columns.str.replace(' ','_',regex=False)
inventory.columns=inventory.columns.str.replace(' ','_',regex=False)
targets.columns=targets.columns.str.replace(' ','_',regex=False)

orders['Date']=pd.to_datetime(orders['Date'])

df=orders.merge(customers,on='CustomerID',how='left')
df=df.merge(products,on='ProductID',how='left')

print(df.head())
print(df.shape)
```

## KPI Summary

```

print("Total Sales:",df['Sales'].sum())
print("Total Profit:",df['Profit'].sum())
print("Total Orders:",df['OrderID'].nunique())
print("Total Customers:",df['CustomerID'].nunique())
print("Average Order Value:",round(df['Sales'].mean(),2))
```

## Monthly Sales

```

monthly=df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum()
monthly.index=monthly.index.astype(str)
monthly.plot(marker='o')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Category Sales

```

cat=df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(cat)
cat.plot(kind='bar')
plt.title('Sales by Category')
plt.tight_layout()
plt.show()
```

## Top Products

```

top=df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top)
top.plot(kind='bar')
plt.title('Top 10 Products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Top Customers

```

cust=df.groupby('Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(cust)
cust.plot(kind='bar')
plt.title('Top 10 Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## State Performance

```

state=df.groupby('State')['Sales'].sum().sort_values(ascending=False)
print(state)
state.plot(kind='bar')
plt.title('Sales by State')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Segment Analysis

```

seg=df.groupby('Segment')[['Sales','Profit']].sum()
print(seg)
seg.plot(kind='bar')
plt.title('Segment Sales & Profit')
plt.tight_layout()
plt.show()
```

## Inventory

```

inventory['Status']=np.where(inventory['Current_Stock']<inventory['Reorder_Level'],'Reorder','Healthy')
print(inventory['Status'].value_counts())
inventory['Status'].value_counts().plot(kind='bar')
plt.title('Inventory Status')
plt.tight_layout()
plt.show()
```

## Correlation

```

print(df[['Sales','Profit','Quantity']].corr())
plt.matshow(df[['Sales','Profit','Quantity']].corr())
plt.colorbar()
plt.title('Correlation Matrix')
plt.show()
```

## Business Insights

```

print("Business Insights")
print("- Highest revenue category:",cat.idxmax())
print("- Highest revenue state:",state.idxmax())
print("- Best customer:",cust.idxmax())
print("- Best product:",top.idxmax())
print("- Products needing reorder:",(inventory['Status']=='Reorder').sum())
```
